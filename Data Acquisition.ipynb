# Data acquisition y data wrangling 

Tenemos dos data set con las mismas columnas, es necesario saber si la información está duplicada o si son dos bases de datos distintas.

```python
import pandas as pd
```

```python
#Leemos el primer data set

df=pd.read_csv('train_strokes.csv',sep=',')
df.head()
```

```python
df.shape
```

```python
df['stroke'].value_counts()
```

```python
# Importamos el segundo data set

df_2=pd.read_csv('healthcare-dataset-stroke-data.csv',sep=',')
```

```python
df_2.head()
```

```python
df_2.shape
```

Ahora los concatenamos y luego se analizarán los datos duplicados a partir del ID 

```python
df_conc=pd.concat([df, df_2],join='outer', ignore_index = True)
df_conc.head()
```

```python
df_conc.shape
```

Al unir ambos registros pasamos a tener un total de 48510 filas. Primero hay que chequear que los valores no sean repetidos, para eso utilizaremos el ID el cual debe ser único por paciente. 

```python
dup_df=df_conc.duplicated(df_conc.columns[~df_conc.columns.isin(['id'])])
```

```python
dup_df.value_counts()
```

Hay 3566 pacientes que se repiten en ambos data set. Solo nos quedaremos con un registro de cada paciente. 

```python
df_no_dup=df_conc.drop_duplicates(keep='first')
df_no_dup.head()
```

```python
df_no_dup.shape
```

```python
df_no_dup['stroke'].value_counts()
```

Transformamos las columnas 

```python
df_rename=df_no_dup.rename(columns={'stroke':'ACV',
                          'smoking_status':'Fumador',
                          'Residence_type':'Residencia', 
                          'work_type':'Trabajo',
                          'ever_married':'Casado',
                          'heart_disease':'Enfermedad cardiaca',
                          'hypertension':'Hipertension',
                          'age':'Edad',
                          'bmi':'IMC',
                          'gender':'Sexo',
                          'avg_glucose_level':'Glucemia'})
```

```python
df_rename.head()
```

```python
df_rename['Edad']=df_rename['Edad'].astype(int)
df_rename.head()
```

Hemos obtenido un data set con más datos y sin valores duplicados. Lo exportamos como un archivo .csv para los próximos trabajos
